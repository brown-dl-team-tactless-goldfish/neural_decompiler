```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt


from transformer.dataprocess import DataLoader, read_vocab_from_csv
current_dir = os.path.abspath('.')
```

```python
c_data_path = f"{current_dir}/../data/leetcode_data_FINAL/C_COMPILED_FILES"
asm_data_path = f"{current_dir}/../data/leetcode_data_FINAL/ASM_COMPILED_FILES"


c_vocab_path = f"{current_dir}/vocab/new_c_vocab.csv"
asm_vocab_path = f"{current_dir}/vocab/new_asm_vocab.csv"


c_vocab = read_vocab_from_csv(c_vocab_path)
asm_vocab = read_vocab_from_csv(asm_vocab_path)



loader = DataLoader(c_data_path, asm_data_path)
_, _, stats = loader.load_data(asm_vocab, c_vocab)

c_token_lengths = stats['c_tokenized_lengths']
asm_token_lengths = stats['asm_tokenized_lengths']
```

```python
# Plot C tokens

plt.hist(c_token_lengths, bins=range(min(c_token_lengths), max(c_token_lengths) + 10, 10), edgecolor='black', color='black')

plt.xlabel('Token Length')
plt.ylabel('Frequency')
plt.title('C File Tokenized Lengths')
```

```python
# Plot ASM tokens

plt.hist(asm_token_lengths, bins=range(min(c_token_lengths), max(c_token_lengths) + 10, 10), edgecolor='black', color='black')

plt.xlabel('Token Length')
plt.ylabel('Frequency')
plt.title('ASM File Tokenized Lengths')
```
